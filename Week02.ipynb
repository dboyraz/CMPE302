```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

v = [3, 4, 5, 2]

s = 3 # scalar value

v = np.array(v) # convert to numpy array so it becomes a Vector

v * s  # multiply vector by a scalar value
```

```python
### !!! ASSIGNMENT
## Create two vectors (2d)
## Plot each vector
## Plot v1 + v2 , v1 - v2, (v1 * 4 + v2/2)

v1 = np.array([4, 5])
v2 = np.array([-3, 6])

v_sum = v1 + v2
v_minus = v1 - v2

v3 = (v1 * 4) + ( v2 / 2)

# Draw the vectors
plt.quiver(0, 0, v1[0], v1[1], angles="xy", scale_units="xy", scale = 1, color="Red") # v1
plt.quiver(0, 0, v2[0], v2[1], angles="xy", scale_units="xy", scale = 1, color="Blue") # v2
plt.quiver(0, 0, v_sum[0], v_sum[1], angles="xy", scale_units="xy", scale = 1, color="Black") # v1 + v2
plt.quiver(0, 0, v_minus[0], v_minus[1], angles="xy", scale_units="xy", scale = 1, color="Green") # v1 - v2
plt.quiver(0, 0, v3[0], v3[1], angles="xy", scale_units="xy", scale = 1, color="Orange") # (v1 * 4 + v2 / 2)

# Set the limits for the plot
plt.xlim([-30, 30])
plt.ylim([-10, 30])

# Set the labels for the plot
plt.xlabel('X')
plt.ylabel('Y')

# Set the labels for the vectors
plt.text(v1[0],v1[1],"v1")
plt.text(v2[0],v2[1],"v2")
plt.text(v_sum[0],v_sum[1],"v1 + v2")
plt.text(v_minus[0],v_minus[1],"v1 - v2")
plt.text(v3[0],v3[1], "v1 * 4 + v2 / 2")

# Show the plot
plt.title("Assignment 1")
plt.show()
```

```python
#Assignment 2
# 1) Come up with three 2D vectors, two of them are orthogonal but neither is orthogonal to the third
# 2) Plot all three vectors -- make sure the axes are square and equal

v1 = np.array([5, 2])
v2 = np.array([-2, 5])
v3 = np.array([4, 4])

print(f"v1.v2 = {np.dot(v1, v2)}")
print(f"v1.v3 = {np.dot(v1, v3)}")
print(f"v2.v3 = {np.dot(v2, v3)}")

# Draw the vectors
plt.quiver(0, 0, v1[0], v1[1], angles="xy", scale_units="xy", scale = 1, color="Red") # v1
plt.quiver(0, 0, v2[0], v2[1], angles="xy", scale_units="xy", scale = 1, color="Blue") # v2
plt.quiver(0, 0, v3[0], v3[1], angles="xy", scale_units="xy", scale = 1, color="Black") # v3

# Axes
plt.xlim([-10, 10])
plt.ylim([-10, 10])

# Set the labels for the plot
plt.xlabel('X')
plt.ylabel('Y')

# Set the labels for the vectors
plt.text(v1[0],v1[1],"v1")
plt.text(v2[0],v2[1],"v2")
plt.text(v3[0], v3[1], "v3")

# Show the plot
plt.title("Assignment 2")
plt.show()
```

```python
# Assignment
# 1) Create 3 matrices: 2x2, 2x2, 3x2
# 2) Compute scalar matrix multiplication
# 3) add all pairs of matrices

m1 = np.array([
    [1, 3],
    [5, 6]
])

m2 = np.array([
    [4, 8],
    [3, 4]
])

m3 = np.array([
    [9, 4],
    [3, 2],
    [6, 2]
])

# Scalar value
scalar = 2

# For "prettier" print
print("m1 * 2")
print(np.array2string(m1 * scalar).replace('[[',' [').replace(']]',']'), "\n")
print("m2 * 2")
print(np.array2string(m2 * scalar).replace('[[',' [').replace(']]',']'), "\n")
print("m3 * 2")
print(np.array2string(m3 * scalar).replace('[[',' [').replace(']]',']'), "\n")

# Can only add matrices with same size
print("m1 + m2")
print(np.array2string(m1 + m2).replace('[[',' [').replace(']]',']'))


```

```python
# Assignment
# Create a 3x3 matrix
# Multiply and show the result for each : 1 - Identity, 2 - Zeros 3 - Transpose

m = np.array([
    [1, 3, 6],
    [2, 4, 0],
    [8, 3, 8]
])

# 3x3 Identity matrix
m_i = np.eye(3)

# 3x3 Zero Matrix
m_z = np.zeros([3, 3])

# Transpose of matrix
m_t = np.transpose(m)

print("m * identity")
print(np.array2string(m @ m_i).replace('[[',' [').replace(']]',']'), "\n")
print("m * Zeros")
print(np.array2string(m @ m_z).replace('[[',' [').replace(']]',']'), "\n")
print("m * Transpose")
print(np.array2string(m @ m_t).replace('[[',' [').replace(']]',']'), "\n")
```

```python
# Inverse of a Matrix

M = np.array([
    [ 4, 2],
    [ 5, 3]
])

np.linalg.inv(M)
```

```python
# Eigenvalue and Eigenvector

M = np.array([
    [4,2,4],
    [ 5,3,5],
    [ 6,5,6]
])

evalue, evector = np.linalg.eig(M)

print(f"Eigenvalue: \n{evalue}\n")

print(f"Eigenvector: \n{evector}")
```
